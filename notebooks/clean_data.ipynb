```python
import pandas as pd
# 把字串轉成數字，無法轉的就變成 NaN
df['Max_BPM'] = pd.to_numeric(df['Max_BPM'], errors='coerce')
```

```python
# 找出 Dtype='float64' 的所有欄位
numeric_cols = df.select_dtypes(include=['float64']).columns
# 利用各欄位的‘平均數’填補缺失值
df[numeric_cols] = df[numeric_cols].fillna(df[numeric_cols].mean())
```
